```python
# Create a subplot figure
fig = make_subplots(
    rows=4, cols=1,
    subplot_titles=('JobRole vs Attrition', 'MaritalStatus vs Attrition', 'JobInvolvement vs Attrition', 'JobLevel vs Attrition')
)

# Define a function to create grouped bar charts
def add_bar_traces(df, category, row, col):
    counts = df.groupby([category, 'Attrition']).size().reset_index(name='count')
    for attrition in df['Attrition'].unique():
        filtered = counts[counts['Attrition'] == attrition]
        fig.add_trace(
            go.Bar(
                x=filtered[category],
                y=filtered['count'],
                name='Left' if attrition == 1 else 'Stayed',  # Labeling based on attrition value
                marker_color='indianred' if attrition == 1 else 'lightsalmon'  # Color based on attrition value
            ),
            row=row, col=col
        )

# Add subplots
add_bar_traces(df, 'JobRole', 1, 1)
add_bar_traces(df, 'MaritalStatus', 2, 1)
add_bar_traces(df, 'JobInvolvement', 3, 1)
add_bar_traces(df, 'JobLevel', 4, 1)

# Update layout
fig.update_layout(
    height=2000,
    title_text="Attrition Analysis",
    barmode='group'
)

# Show figure
fig.show()
```

# 3. Artificial Neural Network Classifier

```python
import tensorflow as tf

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units=500, activation = 'relu', input_shape = (50,)))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
epochs_hist = model.fit(x_train,y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(x_test)

y_pred = (y_pred>0.5)
```

```python
y_pred
```

```python
plt.plot(epochs_hist.history['loss'] )
```

```python
plt.plot(epochs_hist.history['accuracy'] )
```

```python
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```
